# Introduction to Explanatory Data Analysis

Dataset Link: ```https://collegescorecard.ed.gov/data/```

## Dataset Description:

1. The dataset was created for increasing the accessibility of the information about educational institutes among general people. This dataset will help students to prepare themselves for getting admission into their choice of schools. On the other side, it will help the parents to estimate the costs for the schools according to the academic year. Moreover, this dataset also provides information about the success rate based on their salary after the completion of the school.

2. The Data used here was gathered under the College Scorecard Project by the U. S. Department of Education

3. The dataset is composed of various instances from different aspects. The dataset broadly covers instances about students, educational school of different types & faculties. The instances about schools are such as: admission rate, completion rate, location, governing type, gender distribution, program types and costing per academic year. On the other hand, this dataset provides several instances about students such as: earnings, completion time, races, retention rate and so on. Finally, this dataset also provides information about faculty position salary and their differences in different types of schools.

## Selectiveness of the schools based on the SAT score values of the studetns.

### ```Note:``` 
1. Selectiveness means the degree of success of the schools.
2. I have selected SAT score for selectiveness analysis. 
3. My intuition is, most of the competitive schools requires higher SAT score for admission. And, after the analysis it was clear that SAT score can be one of the important factor for selectiveness analysis. 

So, lets begin our analysis for selectiveness

````python
# Fill dataframe NaN values with mean
cleaning_col=['ADM_RATE_ALL','SAT_AVG', 'RET_FT4', 'C150_4', 'COSTT4_A']
selectiveness_df[cleaning_col]=selectiveness_df[cleaning_col].fillna(selectiveness_df[cleaning_col].mean())
````

* In this dataset there is a garbage value as a string ```PrivacySuppressed```. We need to remove this value from the dataset.
* Otherwise, any kind of ploting of numeric number with the concerned column is not possible.

````python
selectiveness_df = selectiveness_df[selectiveness_df.MN_EARN_WNE_P10 != "PrivacySuppressed"]
````

````python
selectiveness_df = selectiveness_df[selectiveness_df.MN_EARN_WNE_P6 != "PrivacySuppressed"]
````

* As the concerned column is saved as Object type in the dataset, I am converting the type as float value because of the ploting function requirment.

````python
selectiveness_df['MN_EARN_WNE_P6']=selectiveness_df['MN_EARN_WNE_P6'].astype(str).astype(float)
selectiveness_df['MN_EARN_WNE_P10']=selectiveness_df['MN_EARN_WNE_P10'].astype(str).astype(float)
````

### Now, we will see the selectiveness of schools based on several groups of SAT scores.
* So, we need to Create a new categorical column based on the SAT score values.
* The column is distributed in 6 bins.

````python
# Binning the SAT Average Overall Score
bins = [0, 800, 1000, 1200, 1400, 1600]
labels = ["<=800", "801-1000", "1001-1200", "1201-1400", "1401-1600"]
selectiveness_df["SAT Range"] = pd.cut(selectiveness_df["SAT_AVG"], bins, labels=labels)
````

## Now, to see the success story of the schools I am going to plot a bar curve which will show the following two factors.
1. It will show the relation between SAT groups vs the mean earning after 6 years of the completion 
2. It will show the relation between SAT groups vs the mean earning after 10 years of the completion 

````python
SAT_goupby = selectiveness_df.groupby("SAT Range", as_index=False).agg({'ADM_RATE_ALL': 'mean', 'RET_FT4': 'mean', 'C150_4': 'mean'
, 'MN_EARN_WNE_P10': 'mean', 'MN_EARN_WNE_P6': 'mean', 'UNITID': 'count', 'COSTT4_A': 'mean'})
````

````python
SAT_School_Count_df = pd.DataFrame({"SAT Range": SAT_goupby["SAT Range"], 
                                    "Count of Schools (%)": SAT_goupby["UNITID"]/SAT_goupby["UNITID"].sum()*100})
````

### The relation between earning and SAT score shows that, the students from higher selective schools are more successful after completion of the program.

````python
# Mean Earnings Vs. Average Overall SAT Range
plt.bar(SAT_goupby["SAT Range"], SAT_goupby["MN_EARN_WNE_P10"], color="blue", alpha=0.7,
        label="10 Yrs after Entry")
plt.bar(SAT_goupby["SAT Range"], SAT_goupby["MN_EARN_WNE_P6"], color="darkblue", alpha=1,
        label="6 Yrs after Entry")

plt.xlabel("SAT Average Overall Range")
plt.ylabel("Mean Earnings")
plt.title("Mean Earnings Vs. Average Overall SAT Range")
plt.legend(loc="best")
plt.show()
````

**So, from the above bar charts it is evident that higher SAT score schools leads to higher earnings after the completion**
